```python
#### Baixar prerreq

import os
import glob

import requests
from bs4 import BeautifulSoup
import pandas as pd

import shutil

import pandas as pd
```

```python
#### Definir variaveis globais


# Caminho da pasta onde estão os arquivos
DATA_DIR = '../data'

# URL alvo
URL = 'https://www.ssp.df.gov.br/dados-por-regiao-administrativa/'
```

```python
def baixar_arquivos():
    resp = requests.get(URL)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, 'html.parser')

    # Busca links para xls/xlsx
    links = soup.find_all('a', href=True)
    xls_links = []
    for a in links:
        href = a['href']
        if href.lower().endswith(('.xls', '.xlsx')):
            full_url = href if href.startswith('http') else requests.compat.urljoin(URL, href)
            xls_links.append(full_url)

    if not xls_links:
        print("Nenhum link XLS ou XLSX encontrado.")
        return

    for link in xls_links:
        fname = os.path.basename(link.split('?')[0])
        xls_path = os.path.join(DATA_DIR, fname)
        csv_fname = os.path.splitext(fname)[0] + '.csv'
        csv_path = os.path.join(DATA_DIR, csv_fname)

        # Baixar o arquivo
        print(f'Baixando {link} → {xls_path}')
        r = requests.get(link)
        r.raise_for_status()
        with open(xls_path, 'wb') as f:
            f.write(r.content)

        # Converter para CSV
        print(f'Convertendo {xls_path} → {csv_path}')
        try:
            df = pd.read_excel(xls_path, engine='openpyxl' if fname.lower().endswith('xlsx') else 'xlrd')
        except ValueError:
            df = pd.read_excel(xls_path)  # fallback
        df.to_csv(csv_path, index=False, encoding='utf-8-sig')
        print(f'Arquivo salvo: {csv_path}')

baixar_arquivos()
```

```python
# Primeiro, carregamos as primeiras 6 linhas apenas para extrair o nome do bairro
header_preview = pd.read_csv('../data/Planaltina/06_PLANALTINA-24.csv', nrows=6, header=None)
bairro = header_preview.iloc[4, 1]  # Linha 5 (índice 4), coluna 2 (índice 1)

# Agora carregamos o restante ignorando as 6 linhas inúteis
df = pd.read_csv('../data/Planaltina/06_PLANALTINA-24.csv', skiprows=6)

# Renomeamos as colunas corretamente
df.columns = ['EIXO', 'NATUREZA', 'TOTAL', 'JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ']

# Removemos linhas inúteis
df = df[df['NATUREZA'].notna()]  # remove vazias
df = df[~df['NATUREZA'].str.upper().str.contains('TOTAL|CRIMES|FONTE|OBS', na=False)]

# Adicionamos a coluna 'BAIRRO' com o valor dinâmico extraído
df['BAIRRO'] = bairro

# Selecionamos apenas as colunas que nos interessam
df = df[['BAIRRO', 'NATUREZA', 'JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
         'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ']]

meses = ["JAN", "FEV", "MAR", "ABR", "MAI", "JUN", "JUL", "AGO", "SET", "OUT", "NOV", "DEZ"]

df[meses] = df[meses].apply(pd.to_numeric, errors='coerce')

# Agrupar os meses em trimestres
df['T1'] = df[['JAN', 'FEV', 'MAR']].sum(axis=1)
df['T2'] = df[['ABR', 'MAI', 'JUN']].sum(axis=1)
df['T3'] = df[['JUL', 'AGO', 'SET']].sum(axis=1)
df['T4'] = df[['OUT', 'NOV', 'DEZ']].sum(axis=1)

# Selecionar apenas as colunas finais desejadas
df = df[['BAIRRO', 'NATUREZA', 'T1', 'T2', 'T3', 'T4']]

# Resetamos o índice
df.reset_index(drop=True, inplace=True)

# Salvamos o resultado
df.to_csv('crimes_limpo.csv', index=False)
```
